```python
import pandas as pd
```

```python
import tensorflow as tf
```

```python
house_dataFrame = pd.read_csv('https://download.mlcc.google.com/mledu-datasets/california_housing_train.csv')
```

```python
house_dataFrame.head()
```

```python
house_dataFrame.describe()
```

```python
house_dataFrame.describe().transpose()
```

```python
x_data = house_dataFrame.drop(['median_house_value'],axis=1)
```

```python
y_val = house_dataFrame['median_house_value']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_data = pd.DataFrame(data=scaler.fit_transform(x_data),
                      columns=x_data.columns,
                      index=x_data.index)
```

```python
x_data.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_val, test_size=0.3, random_state=101)
#I usually prefer random_state as 101, anyone pick any number as you wish.
```

```python
x_data.columns
```

```python
age = tf.feature_column.numeric_column('housing_median_age')
rooms = tf.feature_column.numeric_column('total_rooms')
bedrooms = tf.feature_column.numeric_column('total_bedrooms')
pop = tf.feature_column.numeric_column('population')
households = tf.feature_column.numeric_column('households')
income = tf.feature_column.numeric_column('median_income')
long = tf.feature_column.numeric_column('longitude')
latit = tf.feature_column.numeric_column('latitude')
```

```python
#bucketized columns added to model
buc_long = tf.feature_column.bucketized_column(source_column=long, boundaries=[0.2,0.4,0.6,0.8])
buc_latit = tf.feature_column.bucketized_column(source_column=latit,boundaries=[0.2,0.4,0.6,0.8])
```

```python
feat_cols = [age,rooms,bedrooms,pop,households,income,buc_long,buc_latit]
```

```python
feat_cols
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train,
                                                 y=y_train,
                                                 batch_size=10,
                                                 num_epochs=10000,
                                                 shuffle=True)
```

```python
my_optimizer=tf.train.AdagradOptimizer(learning_rate=0.05)
```

```python
my_optimizer=tf.contrib.estimator.clip_gradients_by_norm(my_optimizer,5.0)
```

```python
model = tf.estimator.DNNRegressor(hidden_units=[16,10,5,5,4],
                                  optimizer=my_optimizer,
                                  feature_columns=feat_cols)
```

```python
model.train(input_fn=input_func,steps=10000)
```

```python
predict_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test,
                                                         batch_size=10,
                                                         num_epochs=1,
                                                         shuffle=False)
```

```python
pred_gen = model.predict(predict_input_func)
```

```python
predictions = list(pred_gen)
```

```python
final_preds=[]
for pred in predictions:
    final_preds.append(pred['predictions'])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,final_preds)**0.5
```
